# Circular Motion

## Angular Velocity

Angular velocity represents the change in angle per time unit (e.g. $\frac{\Delta\theta}{\Delta t}$).

Normally, __angular velocity is represented by the variable $\omega$__ (omega).

__ANGULAR VELOCITY $= \frac{2\pi}{T}$__

## Graphing
Frequently, circular motion is graphed using parametric equations, in which the parameter $t$ represents seconds.

But because __$t$ represents time, not radians or degrees,__ you must convert $t$ in seconds to $\theta$ in radians (or degrees).

#### Converting Time to Radians
In the classroom exercise, the polar form of the function in the example was:
$$\theta(t) = \frac{\pi}{2}t$$

We must convert time $t$ in seconds to radians in order to find the parametric equations that represent the circular motion:

$$x(t) = R\cos(\omega\cdot t) \quad y(t) = R\sin(\omega\cdot t) \quad \omega = \frac{\pi}{2}$$

So, the final __parametric equations__ are:

$$x(t) = R\cos(\frac{\pi}{2}t) \quad y(t) = R\sin(\frac{\pi}{2}t)$$

## Velocity and Acceleration in Circular Motion

__***__ In the class exercise, the velocity of the circular motion was constant, so the __magnitude of the acceleration = 0.__ However, because __velocity and acceleration are vector quantities,__ and because the __VELOCITY DID CHANGE&mdash;IN DIRECTION.__ Therefore, the velocity and acceleration __DID CHANGE:__ but in direction only.

#### Centripetal Acceleration
Means: __Pointing to center__

#### Centrifugal Acceleration
Means: __Pointing away from center__

## Uniform Circular Motion
i.e. circular motion with __constant velocity.__

$$\theta = \omega t \\
\omega = \text{Angular Velocity} \\
\omega = \frac{2\pi}{T} \\ 
T = \text{period} \\
\\
x = R\cos(\theta) = R\cos(\omega t) \\
y = R\sin(\theta) = R\sin(\omega t)
$$

#### Position Vector
$$\vec{r}(t) = x(t)\cdot i + y(t)\cdot j \\
\vec{r}(t) = R\cos(\omega t) \cdot i + R\sin(\omega t) \cdot j \\
\left|\vec{r}(t)\right| = \sqrt{x^2+y^2} = \sqrt{[R\cos(\omega t)]^2 + [R\sin(\omega t)]^2} = \sqrt{R^2\cos^2(\omega t) + R^2\sin^2(\omega t)} = \sqrt{R^2\left(\cos^2(\omega t) + \sin^2(\omega t)\right)}
$$

#### Differentiating to Find Velocity Vectors
$$\frac{d}{dt}\left[\cos(\omega t)\right] = R\left(-\sin(\omega t)\right)\omega = -R\omega\sin(\omega t)$$

$$\frac{d}{dt}\left[\sin(\omega t)\right] = R\omega\cos(\omega t)$$

$$\vec{v} = -R\omega\sin(\omega t)i + R\omega\cos(\omega t)j$$

#### Calculate $\vec{v} \cdot \vec{r}$

$$\vec{v}\cdot\vec{r} = \left|\vec{v}\right|\left|\vec{r}\right|\cos(\vec{v},\vec{r})$$

$$\vec{v}\cdot\vec{r} = \left[R\cos(\omega t) \cdot \left(-R\omega\sin(\omega t)\right)\right] + \left[R\sin(\omega t)\cdot R\omega\cos(\omega t)\right]$$

#### Calculate Acceleration Vector

$$\vec{a} = \frac{d}{dt}\left[\vec{v}\right] = \frac{d}{dt}\left[ -R\omega\sin(\omega t)i + R\omega\cos(\omega t)j \right] = -R\omega^2\cos(\omega t)i - R\omega^2\sin(\omega t)$$

$$\vec{a} = -\omega^2\vec{r}$$
(__Note:__ Acceleration vector is always opposite position vector in circular motion)

__Magnitude of Velocity Vector:__ $\quad\left|\vec{v}\right| = R\omega$<br/>
__Magnitude of Acceleration Vector:__ $\quad\left|\vec{a}\right| = R\omega^2$

$$\omega = \frac{V}{R}$$
