This chapter lays the foundation for understanding motion by teaching you the fundamental _concepts_ of motion and tools for _visualizing_ motion.

## 1.1 Motion Diagrams
<ol class='alpha outline'>
    <li>
        <dl>
            <dt>Definition of Motion</dt><dd>Change in an object's position wrt time.</dd>
        </dl>
        <ul>
            <li>
                <h4>Types of Motion</h4>
                <ul>
                    <li>
                        <p><strong>Translational Motion:</strong> Motion in which an object moves through space.</p>
                        
                        <h6 style='margin: 1rem 0 0 0;'>Types</h6>
                        <dl style='margin-left: 4rem;'>
                            <dt style='display: list-item; list-style-type: circle;'>Linear Motion</dt><dd></dd>
                            <dt style='display: list-item; list-style-type: circle;'>Circular Motion</dt><dd></dd>
                            <dt style='display: list-item; list-style-type: circle;'>Projectile Motion</dt><dd></dd>
                        </dl>
                    </li>
                    <li>
                        <dl>
                            <dt style='display: list-item; list-style-type: disc;'>Rotational Motion</dt><dd>Change in an object's <strong>angular position</strong> wrt time</dd>
                        </dl>
                    </li>
                </ul>
            </li>
        </ul>
    </li>
    
    <li>
        <dl>
            <dt>Motion Diagram</dt>
            <dd>
                <p>A diagram composed of frames, which depict an object's change in position over multiple <em>equally spaced</em> instants of time.</p>
                
                <p>Motion diagrams can depict an object's position changes (or lack thereof&mdash;i.e. rest), constant velocity, and acceleration based on how an object appears in the diagram.</p>
            </dd>
        </dl>
    </li>
</ol>

## 1.2 Particle Model
<ol class='alpha outline'>
    <li>
        <dl>
            <dt>Definition</dt>
            <dd>
                <p>An abstraction to simplify models of motion.</p>
                <p>For types of motion (e.g. balls, cars, rockets) which are not influenced <em>as a whole</em> by the size or shape of the object in motion, you treat the object as a <strong>particle:</strong> an object who's mass is concentrated at a single point&mdash;particles have no size, shape, nor orientation (distinction between top/bottom/front/back).</p>
                <p>In motion diagrams using the particle model, objects are represented as a simple dot.</p>
            </dd>
        </dl>
    </li>
    <li>
        <dl>
            <dt>Usage</dt><dd></dd>
        </dl>
    </li>
</ol>
