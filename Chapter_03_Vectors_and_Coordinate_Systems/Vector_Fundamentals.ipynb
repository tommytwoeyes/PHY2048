# Vectors

## Definition
Some quantities in the world are completely described using a single number (with a unit, of course). These are called __scalar__ quantities.

__Vector__ quantities, in contrast, are those which __have an associated direction,__ and thus cannot be described using a single number.

A _vector_ is a quantity which has both __magnitude__ (size) and __direction.__

## Symbology

Vectors are represented mathematically as variables using symbols with small arrows above them. No matter which direction the vector itself points, _the symbol always has an arrow pointing to the right._

$$\text{Symbol for a vector quantity:} \space \Large\vec{v}$$

The __magnitude__ of a vector can be depicted symbolically in a couple of ways: by a vector symbol within an absolute value sign, or double absolute value sign, or&mdash;more frequently&mdash;by the letter without the little arrow on top.

$$\text{e.g.}\quad v \space = \space |\vec{v}| \space = ||\vec{v}|| \space = \space 5 \space \small{\text{m/s}}$$

## Geometric Representation

The geometric representation of a vector is an arrow, with the tail of the arrow placed _at the point where it the measurement is made._ Vectors are always drawn with an arrowhead at the tip.

<span style='font-weight: bold; color: orange;'>Note:</span> Although the vector arrow is drawn across the page, from its tail to its tip, this does not indicate that the vector "stretches" across this distance. Instead, __the vector arrow tells us the value of the vector quantity only at the one point where the tail of the vector is placed.__

## Rough Document Outline
* Vector magnitude can be positive or zero, but __never negative.__

* __Symbols:__ vectors are represented symbolically as _variables with arrows above them_&mdash;e.g. $\vec{v}$ or $\vec{A}$. Variables without arrows&mdash;e.g.$r$ and $A$&mdash;are symbols for scalar values. __This distinction is very important, because frequently both variables&mdash;e.g. $r$ and $\vec{r}$&mdash;will be used in the same problem, to represent _different quantities.___

* The arrow over the symbol always points to the right.

* Example __vector__ quantities:
    * Displacement
    * Velocity
    * Acceleration
    * Force
    
* Example __scalar__ quantities:
    * Time
    * Temperature
    * Mass
    * Energy
    * Pressure
